## Representar las funciones de coseno y seno a la vez

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
```

```python
x
```

```python
# visualización
plt.plot(x, np.cos(x))

# añadimos etiquetas y título
plt.xlabel("x")
plt.ylabel("cos(x)")

# Mostramos los resultados
plt.show()
```

```python
#representación de seno y coseno cambiando el tipo de línea
plt.plot(x, np.cos(x), 'r+',x, np.sin(x),'bo') # red plus - blue dot
plt.xlabel("x = cos(x)")
plt.ylabel("y = sin(x)")
plt.legend()
plt.show()
```

# Histogramas

```python
# Parámetros de la función gaussiana
mu, sigma = 50, 7

# generamos un array
x = mu + sigma * np.random.randn(10000)

# la función hist() calcula la frecuencia y el número de barras
n, bins, patches = plt.hist(x, 30, density=1,
                           facecolor='dodgerblue',
                           alpha=0.5)
plt.xlabel('x')
plt.ylabel('Probabilidad')
plt.text(20, .025, r'$\mu=50,\ \sigma=7$')

# Controlamos el tamaño de los ejes (xmin, xmax, ymin, ymax)
plt.axis([10, 90, 0, 0.07])

# Mostramos el grid
plt.grid(True)

# Mostramos el resultado
plt.show()
```

### Agrupaciones con pd.qcut() y pd.cut()

```python
# Creamos una agrupación en n grupos con pd.qcut()
titanic['AgeGroup_qcut'] = pd.qcut(titanic.Age, 5, labels=['A','B','C','D','E'])
```

```python
sns.barplot(x='AgeGroup_qcut',y='Survived', data=titanic)
```

```python
# creamos una agrupación en n grupos con pd.cut()
bins = [-np.inf, 12, 24, 35, 60, np.inf]
labels = ['Unknow', 'Teenager', 'YoungAdult', 'Adult', 'Senior']
titanic['Age'] = titanic['Age'].fillna(-0.5) # imputamos -0.5 a los valores nulos
titanic['AgeGroup_cut'] = pd.cut(titanic.Age, bins, labels = labels)
```

```python
sns.barplot(x="AgeGroup_cut", y='Survived', data=titanic)
```

```python
# también un extremo
bins = [-1, 0, 5, 12, 18, 24, 35, 60, np.inf]
labels = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Senior']
titanic['AgeGroup_cut_2'] = pd.cut(titanic["Age"], bins, labels = labels)
```

```python
sns.barplot(y="AgeGroup_cut_2", x='Survived', data=titanic, palette="flare")
```
